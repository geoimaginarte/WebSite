## Combinar los distintos csv

```python
import glob, os
 
os.chdir("E:/Trabajos Fran/Prospectiva investigación de mercado/TopTalent_Grecia/Data/rutas/a_universal")
results = pd.DataFrame([])
 
for counter, file in enumerate(glob.glob("Top Talent*")):
    namedf = pd.read_csv(file, skiprows=0, usecols=[0,1])
    results = results.append(namedf)
 
results.to_csv('E:/Trabajos Fran/Prospectiva investigación de mercado/TopTalent_Grecia/Data/rutas/a_universal/combinedfile.csv')
```
